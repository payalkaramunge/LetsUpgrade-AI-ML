# LOADING DATASET

```python
import pandas as pd
ds = pd.read_csv('C:/Users/Legion/Desktop/Basvaprasad/AI-ML/DAY 24/train.csv')
ds.head()
```

```python
ds.shape
```

# DROPPING FEATURES

```python
ds.drop(['PassengerId','Name','Ticket','Cabin','Fare','Age'],axis=1,inplace=True)
ds.head()
```

# CONVERTING TEXT TO NUMBER 

```python
from sklearn.preprocessing import LabelEncoder
ec = LabelEncoder()
ds['Sex'] = ec.fit_transform(ds['Sex'])
ds['Embarked'] = ec.fit_transform(ds['Embarked'])
ds.head()
```

# APPYLING SVM

```python
col = list(ds)

for i in col:
    
    y = ds[i]
    X = ds.drop(i,axis=1)
    
    from sklearn.model_selection import train_test_split
    X_train , X_test, y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=0)
    
    from sklearn import svm  
    from sklearn.metrics import accuracy_score    
    from sklearn.metrics import confusion_matrix   
    
    clf = svm.SVC(gamma=0.01,C=110)
    y_perd = clf.fit(X_train,y_train).predict(X_test)
    
    print('Dependent Variable:',i)
    print(confusion_matrix(y_test,y_perd))
    print('Accuracy of Model:',accuracy_score(y_test,y_perd,normalize=True),'\n')
```

# APPYLING KNN

```python

# col = list(ds)
a=1
while (a < 10):
    for i in col:
    
        y = ds[i]
        X = ds.drop(i,axis=1)
    
        from sklearn.model_selection import train_test_split
        X_train , X_test, y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=0)
    
        from sklearn import neighbors  
        from sklearn.metrics import accuracy_score    
        from sklearn.metrics import confusion_matrix   
    
        knn = neighbors.KNeighborsClassifier(n_neighbors=a)
        y_perd = knn.fit(X_train,y_train).predict(X_test)
        
        print('n_neighbors:',a,'\n')
        print('Dependent Variable:',i)
        print(confusion_matrix(y_test,y_perd))
        print('Accuracy of Model:',knn.fit(X_train,y_train).score(X_test,y_test),'\n')
    a+=1
```
